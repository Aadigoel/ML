```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
data = pd.DataFrame({'Temperature':[30,35,40,45,50,55,60,65,70,75], 'Resistance':[4.82,4.84,4.92,4.94,4.99,5.00,5.09,5.11,5.14,5.19]})
data
```

```python
x = data[['Temperature']]
y = data[['Resistance']]
```

```python
regression_model = LinearRegression()
regression_model.fit(x,y)
```

```python
y_predicted = regression_model.predict(x)
print('Slope: ',regression_model.coef_)
print('Intercept: ', regression_model.intercept_)
```

```python
plt.figure(figsize=(3,2))
plt.scatter(x,y, s=10)
plt.title("Resistance vs Temperature")
plt.xlabel("Temperature in degrees Celsius")
plt.ylabel("Resistance in Ohms")
plt.plot(x, y_predicted, color='r')
plt.grid(True)
plt.show()
print("Scale:\nX-axis: 1unit = 10degrees C\nY-axis: 1unit = 0.02V")
```

```python
data1 = pd.DataFrame({'Temperature': [363,358,353,348,348,338,333,328,323,318], 'Junction Voltage': [1.523,1.539,1.548,1.552,1.565,1.577,1.580,1.589,1.608,1.613]})
data1
```

```python
x1 = data1[['Temperature']]
y1 = data1[['Junction Voltage']]
```

```python
regression_model1 = LinearRegression()
regression_model1.fit(x1,y1)
```

```python
y_predicted1 = regression_model1.predict(x1)
print('Slope: ',regression_model1.coef_)
print('Intercept: ', regression_model1.intercept_)
```

```python
plt.figure(figsize=(3,2))
plt.scatter(x1,y1, s=10)
plt.plot(x1, y_predicted1, color='r')
plt.xlabel("Temperature in K")
plt.ylabel("Junction voltage in V")
plt.title("Junction voltage vs Temperature")
plt.grid(True)
plt.show()
print("Scale:\nX-axis: 1unit = 10K\nY-axis: 1unit = 0.02V")
```
